```python
# ì†ìµê³„ì‚°ì„œì˜ ë§¤ì¶œì•¡, ë§¤ì¶œì›ê°€, íŒê´€ë¹„, ì˜ì—…ì†ìµë§Œ ì¶”ì¶œ

# ì¬ë¬´ì œí‘œ ìš”ì²­ URL
BASE_URL = "https://opendart.fss.or.kr/api/fnlttSinglAcntAll.json"

def get_filtered_financial_statement(corp_code: str, year: str, fs_div: str = "CFS"):
    """
    ì†ìµê³„ì‚°ì„œì—ì„œ ë§¤ì¶œì•¡, ë§¤ì¶œì›ê°€, íŒê´€ë¹„, ì˜ì—…ì†ìµë§Œ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜
    """
    params = {
        "crtfc_key": API_KEY,
        "corp_code": corp_code,
        "bsns_year": year,
        "reprt_code": "11011",  # 11011: ì‚¬ì—…ë³´ê³ ì„œ
        "fs_div": fs_div
    }

    # API ìš”ì²­
    response = requests.get(BASE_URL, params=params)
    data = response.json()

    if data['status'] != '000':
        print(f"API ì˜¤ë¥˜: {data['message']}")
        return None

    # JSON ë°ì´í„°ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
    reports = data['list']
    df = pd.DataFrame(reports)

    # ì†ìµê³„ì‚°ì„œ í•­ëª© í•„í„°ë§ ì¡°ê±´
    target_accounts = ["ë§¤ì¶œì•¡", "ë§¤ì¶œì›ê°€", "íŒë§¤ë¹„ì™€ê´€ë¦¬ë¹„", "ì˜ì—…ì´ìµ"]

    # í•„ìš”í•œ ì»¬ëŸ¼ë§Œ ì¶”ì¶œ
    if not df.empty:
        df = df[df['account_nm'].isin(target_accounts)]
        df = df[['sj_nm', 'account_nm', 'thstrm_amount', 'frmtrm_amount']]

        # ì»¬ëŸ¼ ì´ë¦„ ë§¤í•‘
        df.columns = ['ì¬ë¬´ì œí‘œëª…', 'ê³„ì •ëª…', 'ë‹¹ê¸° ê¸ˆì•¡', 'ì „ê¸° ê¸ˆì•¡']

        # ê²°ì¸¡ê°’ ë° ë°ì´í„° ì •ì œ
        df.replace('', '0', inplace=True)
        df[['ë‹¹ê¸° ê¸ˆì•¡', 'ì „ê¸° ê¸ˆì•¡']] = df[['ë‹¹ê¸° ê¸ˆì•¡', 'ì „ê¸° ê¸ˆì•¡']].apply(pd.to_numeric)

    return df

# ì˜ˆì œ ì‹¤í–‰: ì‚¼ì„±ì „ì(00126380) 2022ë…„ ì†ìµê³„ì‚°ì„œ ì£¼ìš” í•­ëª© ì¶”ì¶œ
corp_code = "00126380"  # DARTì—ì„œ ì œê³µí•˜ëŠ” íšŒì‚¬ ì½”ë“œ
year = "2022"
filtered_financial_data = get_filtered_financial_statement(corp_code, year)

if filtered_financial_data is not None:
    print(filtered_financial_data)
    # ì—‘ì…€ë¡œ ì €ì¥
    filtered_financial_data.to_excel("income_statement_filtered.xlsx", index=False)
```

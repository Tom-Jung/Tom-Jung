```
import requests
from bs4 import BeautifulSoup
import openpyxl
import time
from datetime import datetime

# 삼성전자(005930) 주가 정보 URL
url = "https://finance.naver.com/item/main.nhn?code=005930"
headers = {"User-Agent": "Mozilla/5.0"}

# 매입가 설정
purchase_price = 72000

# 엑셀 파일 생성 또는 불러오기
file_name = "real_time_stock.xlsx"

try:
    wb = openpyxl.load_workbook(file_name)
    sheet = wb.active
except FileNotFoundError:
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet["A1"] = "시간"
    sheet["B1"] = "현재 주가 (KRW)"
    sheet["C1"] = "매입가 (KRW)"
    sheet["D1"] = "손익 (KRW)"

# 매입가 고정
sheet["C2"] = purchase_price

while True:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    current_price = int(soup.select_one(".no_today .blind").text.replace(",", ""))  # 쉼표 제거 후 숫자로 변환

    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # 특정 셀(A2, B2, D2) 업데이트
    sheet["A2"] = now
    sheet["B2"] = current_price
    sheet["D2"] = current_price - purchase_price  # 손익 계산

    # 엑셀 저장
    wb.save(file_name)

    print(f"{now} - 현재 주가: {current_price} KRW | 매입가: {purchase_price} KRW | 손익: {current_price - purchase_price} KRW")

    time.sleep(1)  # 1초마다 업데이트
```
